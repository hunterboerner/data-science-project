```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt# Load data
from sklearn import linear_model

%matplotlib inline
city_data = pd.read_csv('USData_ClassProject1.csv')
```

```python
clf = linear_model.LinearRegression()
clf.fit(city_data['PercentOver65'],city_data['AvgHighTemp'])

plt.scatter(city_data.PercentOver65,city_data.AvgHighTemp,color="Black")
plt.xlabel('AvgHighTemp')
plt.ylabel('PercentOver65')
```
